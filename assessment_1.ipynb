```python
#open image
import numpy as np
import matplotlib.pyplot as plt
import cv2
img=cv2.imread('cat-4.jpg')
plt.imshow(img)
```

```python
import cv2
img_rgb= cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
new_img=img_rgb.copy()
new_img=cv2.flip(new_img,0)
plt.imshow(new_img)
```

```python
pt1=(200,450)
pt2=(800,200)
cv2.rectangle(img_rgb,pt1=pt1,pt2=pt2,color=(255,0,0),thickness=3)
plt.imshow(img_rgb)
img_rgb= cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
vertices=np.array([[250,700],[425,400],[600,700]],np.int32)
vertices.shape
```

```python
pts=vertices.reshape((-1,1,2))
vertices
```

```python
pts.shape
```

```python
cv2.polylines(img_rgb,[pts],isClosed=True,color=(0,0,255),thickness=20)
plt.imshow(img_rgb)
```

```python
import cv2
img_rgb= cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
cv2.fillPoly(img_rgb,[pts],color=(0,0,255))
plt.imshow(img_rgb)
```

```python
import cv2
import numpy as np
def create_circle(event,x,y,flag,param):
    if event==cv2.EVENT_RBUTTONDOWN:
        cv2.circle(img,(x,y),100,(0,0,255),thickness=10)
img=cv2.imread("cat-4.jpg")
cv2.namedWindow(winname='cat')
cv2.setMouseCallback('cat',create_circle)
while True:
    cv2.imshow('cat',img)
    if cv2.waitKey(20) & 0xFF==27:
        break
cv2.destroyAllWindows()
    
```
