```python
import cv2
image = cv2.imread('cat-3.jpg')
rotat_90 = cv2.rotate(image, cv2.ROTATE_90_CLOCKWISE)
rotat_180 = cv2.rotate(image, cv2.ROTATE_180)
rotat_270 = cv2.rotate(image, cv2.ROTATE_90_COUNTERCLOCKWISE)
cv2.imshow('Rotate 90', rotat_90)
cv2.imshow('Rotate 180', rotat_180)
cv2.imshow('Rotate 270', rotat_270)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
